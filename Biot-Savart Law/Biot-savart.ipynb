```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
from scipy.optimize import curve_fit
import pandas as pd
df= pd.read_csv('B_vs_I.dat')
plt.style.use(['science','no-latex','grid'])
I=df.I
B_120=df.B_120
```

```python
B=[]
for i in B_120:
    B.append(i/10)
```

```python
B_120T=[]
u=4*np.pi*10**-3
for i in range(0,21,1):
    B_120T.append(i*u/(2*(60*10**-3)))
B_120T
```

```python
Ir=range(0,21,1)
plt.plot(Ir,B_120T,'k-',label='theory')
plt.plot(I,B,'r.-',label='r=120mm')
plt.legend()
plt.show()
```

```python
d=np.arange(-10,11,1)
i=20
u=4*np.pi*10
R=20*10^-3
s=R**2
Bx=[]
for x in d:
    z=2*(R**2+(x*10^-3)**2)
    if x==0:
        Bx.append(8)
    else:
        Bx.append(i*u*s/(z**1.5))

plt.plot(d,Bx,'b-')
plt.show()
```
