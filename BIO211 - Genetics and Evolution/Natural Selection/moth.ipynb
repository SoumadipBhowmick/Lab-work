```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','no-latex'])
df=pd.read_csv('moth_data_10_10_2023_Tue_Batch.xlsx - Sheet1.csv',delimiter=',')
df
```

```python
X = ['Trial 1','Trial 2','Trial 3'] 
p_light = [0,8,33] 
p_dark =[100,92,67] 
  
X_axis = np.arange(len(X)) 
plt.figure(figsize=(6,5))
plt.bar(X_axis - 0.2, p_light, 0.4, label = '% Light',color='white',edgecolor='black',
       linewidth=1) 
plt.bar(X_axis + 0.2, p_dark, 0.4, label = '% Dark', color='black') 
  
plt.xticks(X_axis, X) 
plt.xlabel("Trials") 
plt.ylabel("Moth morphs (Percentage)") 
plt.title("Percentage of moth morps remaining") 
plt.legend() 
plt.savefig('soumo.png',dpi=300)
plt.show() 
```

```python
x=np.array(df['LM'])
y=np.array(df['LS'])
y
```

```python
x[62]='0'
y[62]='0'
```

```python
bg=df['BG']
dm=df['DM']
ds=df['DS']
lm=x
ls=y

dm=dm.astype(np.float64)
ds=ds.astype(np.float64)
lm=lm.astype(np.float64)
ls=ls.astype(np.float64)
```

```python
limov=[]
lista=[]
damov=[]
dasta=[]
limov_da=[]
lista_da=[]
damov_da=[]
dasta_da=[]
for i in range(len(bg)):
    if bg[i] == 'Light Forest':
            limov.append(lm[i])
            lista.append(ls[i])
            damov.append(dm[i])
            dasta.append(ds[i])
    if bg[i] == 'Dark Forest':
            limov_da.append(lm[i])
            lista_da.append(ls[i])
            damov_da.append(dm[i])
            dasta_da.append(ds[i])
```

```python
lifor=[limov,lista,damov,dasta]
plt.figure(figsize=(7,6))
plt.title('Distribution of moths in Light Forest')
plt.boxplot(lifor,labels=['Light moving','Light Stationary','Dark Moving', 'Dark Stationary'])
plt.show()
```

```python
dafor=[limov_da,lista_da,damov_da,dasta_da]
plt.figure(figsize=(7,6))
plt.title('Distribution of moths in Dark Forest')
plt.boxplot(dafor,labels=['Light moving','Light Stationary','Dark Moving', 'Dark Stationary'])
plt.show()
```
