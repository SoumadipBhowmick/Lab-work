```python
l=[19.6,22.5,25.7,29,31.6]
M=[1,1.5,2,2.5,3]
M=np.array(M)

from scipy.optimize import curve_fit
def f(x,a,b):
    return a*x+b
popt,pcov =curve_fit(f,np.sqrt(M*1e3),l)
x=np.linspace(30,60,num=100)
M=np.array(M)
#
plt.figure(figsize=(5, 4))


plt.plot(x,f(x,*popt),'r-')
plt.plot(np.sqrt(M*1e3),l,'ko')
plt.xlabel('Root Mass ($\sqrt{g}$)')
plt.ylabel('Length (cm)')
```
