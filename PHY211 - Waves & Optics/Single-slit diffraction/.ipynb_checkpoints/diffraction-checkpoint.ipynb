```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','no-latex'])
```

```python
X=np.linspace(-0.05,0.05,num=500)
a=7*1e-5
l=623.8*1e-9
b=2*np.pi*a*np.sin(X)/l
Y=(np.sin(b/2)/(b/2))**2
```

```python
plt.figure(figsize=(6,5))
plt.axhline(y=0)
plt.plot(X,Y,'r--')
plt.xlabel('$/theta$')
plt.ylabel('$Intensity (I(/theta))$')
```
