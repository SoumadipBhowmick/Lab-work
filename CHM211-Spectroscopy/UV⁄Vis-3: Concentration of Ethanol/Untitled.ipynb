```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','no-latex'])
df=pd.read_csv('UVis-3.csv',delimiter=',')
wave=np.array(df['Wavelength (nm)'])


Blank=np.array(df['Blank'])
Abs_1=np.array(df['Sample 1'])
Abs_2=np.array(df['Sample 2'])
Abs_3=np.array(df['sample 3'])
Abs_4=np.array(df['Sample 4'])
Abs_5=np.array(df['Sample 5'])
Abs_6=np.array(df['Sample 6'])
```

```python
plt.figure(figsize=(10,6))
plt.plot(wave[:250],Blank[:250]-Blank[:250],'r--')
plt.plot(wave[:250],Abs_1[:250]-Blank[:250],'b-')
plt.plot(wave[:250],Abs_2[:250]-Blank[:250],'m-')
plt.plot(wave[:250],Abs_3[:250]-Blank[:250],'y-')
plt.plot(wave[:250],Abs_4[:250]-Blank[:250],'c-')
plt.plot(wave[:250],Abs_5[:250]-Blank[:250],'g-')
plt.plot(wave[:250],Abs_6[:250]-Blank[:250],'k-')
```
