```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#import data
Ic,T =np.loadtxt('Ic_T.dat', delimiter=',',unpack=True)
#Ih vs Torque
#plot graph
plt.plot(Ic,T,'bo',label='data')

#fit line
def f(x,a,b):
    return a*x+b

popt, pcov =curve_fit(f,Ic,T)

xFit=np.arange(0.5,2.99,0.00001)
plt.plot(xFit,f(xFit,*popt),'r-',label='line fit')
plt.xlabel('$I_c$ (A)')
plt.ylabel('$Torque$ (Nm)')
plt.legend(loc='lower right')
plt.grid()
plt.show()

```
