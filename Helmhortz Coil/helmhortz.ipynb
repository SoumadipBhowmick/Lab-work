```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#import data
d1=np.loadtxt('Ic_T.dat')
#Ih vs Torque
#plot graph
Ic=d1[:,0]
T=d1[:,1]
plt.clf()
plt.plot(Ic,T,'bo')

#fit line
def f(x,a,b):
    return a*x+b

popt, pcov =curve_fit(f,Ic,T)

xFit=np.arange(0.5,3.5,0.5)
plt.plot(xFit,f(xFit,*popt),'r-')
plt.xlabel('I_c(A)')
plt.ylabel('Torque(Nm)')
plt.grid()
plt.show()

slope, intercept = np.polyfit(xFit,T,1)
print('Slope: ',slope, 'Interept: ',intercept)
```
