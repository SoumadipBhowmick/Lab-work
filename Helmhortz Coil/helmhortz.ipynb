```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#import data
Ic,T =np.loadtxt('Ic_T.dat', delimiter=',',unpack=True)
#Ic vs Torque
#plot graph
plt.plot(Ic,T,'bo',label='data')

#fit line
def f(x,a,b):
    return a*x+b

popt, pcov =curve_fit(f,Ic,T)

xFit=np.arange(0.5,2.99,0.001)
plt.plot(xFit,f(xFit,*popt),'r--',label='line fit')
plt.xlabel('$I_c$ (A)')
plt.ylabel('$Torque$ (Nm)')
plt.legend(loc='lower right')
plt.grid()
plt.show()

```

```python
#I_h vs Torque

Ih,T2=np.loadtext('Ih_T.dat', delimiter=',', unpack=True)

plt.plot(Ih,T2,'bo', label='data')
def f(x,a,b):
    return a*x+b

popt, pcov =curve_fit(f,Ih,T2)

xFit=np.arange(min(Ih),max(Ih),0.00001)
plt.plot(xFit,f(xFit,*popt),'y--',label='line fit')
plt.xlabel('$I_h$ (A)')
plt.ylabel('$Torque$ (Nm)')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
