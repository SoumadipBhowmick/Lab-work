```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#import data
Ic,T =np.loadtxt('Ic_T.dat', delimiter=',',unpack=True)
#Ic vs Torque
#plot graph
plt.plot(Ic,T,'bo',label='data',markersize='4')

#fit line
def f(x,a,b):
    return a*x+b

popt, pcov =curve_fit(f,Ic,T)

xFit=np.arange(0.25,3.3,0.001)
plt.plot(xFit,f(xFit,*popt),'r--',label='line fit')
plt.xlabel('$I_c$ (A)')
plt.ylabel('$Torque$ (Nm)')
plt.xticks(np.arange(0,3.75,0.5))
plt.yticks(np.arange(0,3.75,0.5))
plt.legend(loc='lower right')
plt.grid(linestyle='dotted')
plt.savefig('IcvsT.png',dpi=200)
plt.show()

```

```python
popt
```

```python
m=popt[0]      #popt shows slope and intercept of fitted line
Ih=3           #pcov diagonal shows the error in slope and intercept
r=0.045
A=3.1416*r**2
#m=c*Ih*A
c=m/(Ih*A)
print(c)
```

```python
#I_h vs Torque

Ih,T2=np.loadtxt('Ih_T.dat', delimiter=',', unpack=True)

plt.plot(Ih,T2,'bo', label='data',ms='3')
def f(x,a,b):
    return a*x+b

popt, pcov =curve_fit(f,Ih,T2)

xFit=np.arange(min(Ih),max(Ih),0.00001)
plt.plot(xFit,f(xFit,*popt),'y--',label='line fit')

plt.xticks(np.arange(0,3.75,0.5))
plt.yticks(np.arange(0,3.75,0.5))
plt.xlabel('$I_h$ (A)')
plt.ylabel('$Torque$ (Nm)')
plt.legend(loc='lower right')
plt.grid(linestyle='dotted')

plt.savefig('IhvsT.png',dpi=200)
plt.show()
```

```python
#sinA vs Torque
A,T3=np.loadtxt('A_T.dat',delimiter=',',unpack=True)
As=[]


for i in A:
    j=(i*np.pi)/180

    As.append(np.sin(j)) # i in rad

print(As)

plt.plot(As,T3,'bo',label='data',markersize='4')
def f(x,a,b):
    return a*x+b

popt, pcov =curve_fit(f,As,T3) 

xFit=np.arange(min(As),max(As),0.00001)
plt.plot(xFit,f(xFit,*popt),'g--',label='line fit')
plt.xlabel('$sinA$')
plt.ylabel('$Torque$ (Nm)')
plt.xticks(np.arange(0,1.5,0.25))
plt.yticks(np.arange(0,3,0.25))
plt.legend(loc='lower right')
plt.grid(linestyle='dotted')
plt.savefig('sinAvsT.png',dpi=200)
plt.show()
popt
```
