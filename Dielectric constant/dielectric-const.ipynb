```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#import data
```

```python
#Part A k(air)
#calculate Q
Q=[]
C=220
V=[0,0.6,1.0,1.6,1.9,3.0,3.4,3.8,4.5]
Uc=[0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]
for i in V:
    Q.append(C*i)


# Plot Q vs Uc
def f(x,a,b):
    return a*x+b

plt.plot(Uc,Q,'bo')

popt, pcov =curve_fit(f,Uc,Q,p0=[220,0])

xFit=np.arange(0.0,4.0,0.00001)

plt.plot(xFit,f(xFit,*popt),'r-')
plt.xlabel('Uc (kV)')
plt.ylabel('Q (nAs)')
plt.grid()
plt.show()
```

```python
popt[0]
```

```python
#Calculaitng dielectric const. for air

# As Q=keAUc/d
slope=popt[0]
A=0.05513
d=0.002
e=8.854
#m=keA/d
#k=md/A
k=(slope*d)/(A)

print('k(air): ',k)
```

```python
#Part B Uc vs 1/d

Q=[]
D_inv=[]
C=220
d=[0.1,0.15,0.2,0.25,0.3,0.35,0.4]
V=[3.2,2.4,1.7,1.4,1.2,1.1,1.0]
Uc=[0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]
for i in V:
    Q.append(C*i)
for i in d:
    D_inv.append(1/i)

# Plot Q vs Uc
def f(x,a,b):
    return a*x+b

plt.clf()
plt.plot(D_inv,Q,'bo')

popt, pcov =curve_fit(f,D_inv,Q,p0=[75,45])

xFit=np.arange(0.0,10,0.1)

plt.plot(xFit,f(xFit,*popt),'r-')
plt.xlabel('1/d (cm^-1)')
plt.ylabel('Q (nAs)')
plt.show()

slope2, intercept = np.polyfit(D_inv,Q,1)
print('Slope: ',slope2, 'Intercept: ',intercept)
```
